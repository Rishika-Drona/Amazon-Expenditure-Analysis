```python
import pandas as pd
# readinf the file into the dataframe
df = pd.read_csv("C:/Users/rishi/Desktop/amazon-orders.csv")
df.head()
```

```python
#  to find the full size of the data set with rows and columns
df.shape
```

```python
# Replacing the "NaN" with 0
df = df.fillna(0)
df.head()
```

```python
#CLeaning the data by replacing the '$' with nothing and converitng it to float for easier calculation
df["Item Total"] = df['Item Total'].str.replace('$','').astype(float)
df.head()
```

```python
# total expenditure
df["Item Total"].sum()
```

```python
# average purchased price
df["Item Total"].mean()
```

```python
#to find the outliers
df["Item Total"].median()
```

```python
#Maximum Spent
df["Item Total"].max()
```

```python
# Minimum Spent
df["Item Total"].min()
```

```python
#CLeaning the data by replacing the '$' with nothing and converitng it to float for easier calculation
df["Item Subtotal Tax"] = df['Item Subtotal Tax'].str.replace('$','').astype(float)
df.head()
```

```python
df["Item Subtotal Tax"].sum()
```

```python
# To analyse the overall tax paid from the bought items
df["Item Subtotal Tax"].sum()/df["Item Total"].sum()
```

```python
# Converting the the date into datetime data type so that it becomes easy for the systemt o recognize them as dates
df["Order Date"] = pd.to_datetime(df["Order Date"])
df.head()
```

```python
# Analysis of expenditure on each particular day in dataset by using matplotlib package which will allow us to display charts in the notebook.
%matplotlib inline
```

```python
# Passing the desired X and Y axis columns to the above function as an argument and adding other argument like to rotate the x- axis labels
df.plot.bar(x="Order Date", y="Item Total", rot = 90)
```

```python
# To create a bigger size of the above figure to make it more visible by passing an argument. It allows to define the size of the graph by width and size
df.plot.bar(x="Order Date", y="Item Total", rot = 90, figsize=(20,10))
```

```python
# To get the same bar for the orders placed on the same date since we need to observe the total spent on particular day
daily_orders = df.groupby("Order Date").sum()["Item Total"]
daily_orders.head()
```

```python
#The results of "head()" above looks a bit different here. Thatâ€™s because this is a Series rather than a DataFrame. It looks like it has two columns, but it actually only has one: the 'Order Date' column entries are actually the index labels for our sums.These numbers are index labels. So converting them into bar graph would be as follows:
daily_orders.plot.bar(figsize=(20,10))
```

```python
# to change the color of the bars for example green
daily_orders.plot.bar(figsize=(20,10), color="#61D199")
```
